```python
import pandas as pd
from pycaret.clustering import *
dataset = pd.read_csv("../../datasets/clustering/marketing_campaign.csv", sep="\t")

dataset.head()
```

```python
for cat in dataset:
    if dataset[cat].dtype == "object":
        map = dict(zip(dataset[cat].unique(), range(dataset[cat].unique().size)))
        dataset[cat] = dataset[cat].map(map)

dataset.shape
```

### _setup_

This function initializes the training environment and creates the transformation pipeline. Setup function must be called before executing any other function.

```python
s = setup(dataset, normalize = True)
```

### _create\_model_

This function trains and evaluates the performance of a given model.

```python
kmeans = create_model('kmeans')
print(kmeans)
```

### _evaluate\_model_

This function analyzes the performance of a trained model.

```python
evaluate_model(kmeans)
```

### _assign\_model_

This function assigns cluster labels to the training data, given a trained model.

```python
result = assign_model(kmeans)
result.head()
```
